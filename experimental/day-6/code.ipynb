```python
import numpy as np, matplotlib.pyplot as plt
T = np.logspace(-3, 4, 100)
for rho in [10, 100, 1000]:
    z = 250 * np.sqrt(rho * T)
    plt.loglog(T, z, label=f'{rho} Ω·m')
plt.xlabel('Period (s)')
plt.ylabel('Approx depth of investigation (m)')
plt.legend(); plt.grid(True, which='both')
plt.title('Depth of investigation vs Period')
plt.show()
```
